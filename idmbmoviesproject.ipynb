```python
import pandas as pd

# Load the dataset
r = pd.read_csv('/content/drive/MyDrive/IMDB-Movie-Data.csv')

# Split the genres and create a list of all genres
all_genres = []
for genres in r['Genre']:
    genre_list = genres.split(',')
    all_genres.extend(genre_list)

# Create a Series to count the occurrences of each genre
genre_counts = pd.Series(all_genres).value_counts()

# Display the counts of different genres
print("Genre Counts:")
print(genre_counts)
```

```python
#Present the distribution of movies across different countries using a bar plot.
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
r1 = pd.read_csv('/content/drive/MyDrive/movies_metadata.csv')

# Define a function to extract country names from the improperly formatted data
def extract_countries(row):
    countries = []
    try:
        data = eval(row)  # Convert the row string to a Python object
        if isinstance(data, list):
            for item in data:
                if isinstance(item, dict) and 'name' in item:
                    countries.append(item['name'])
    except:
        pass
    return countries

# Apply the extraction function to the 'production_countries' column
r1['production_countries'] = r1['production_countries'].apply(extract_countries)

# Flatten the list of country names
countries_list = [country for countries in r1['production_countries'] for country in countries]

# Create a pandas Series to count country occurrences
country_counts = pd.Series(countries_list).value_counts()

# Plotting the bar plot
plt.figure(figsize=(12, 6))
country_counts[:20].plot(kind='bar', edgecolor='k', alpha=0.7)
plt.title('Distribution of Movies Across Countries')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```
